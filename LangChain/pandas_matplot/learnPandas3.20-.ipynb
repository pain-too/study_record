```python
#第一道例题
import pandas as pd
import numpy as np

np.random.seed(42)
values = np.random.randint(50,101,10)
index = []
for i in range(1,11):
    index.append('学生'+str(i))
scores = pd.Series(values,index)
scores
```

```python
#案例一
import pandas as pd
data = {
    '姓名':['张三','李四','王五','赵六','钱七'],
    '数学':[85,92,78,88,95],
    '英语':[90,88,85,92,80],
    '物理':[75,80,88,85,90]
}

scores = pd.DataFrame(data)#把字典直接看作dataframe类型

#计算每个学生的总分
scores['总分']=scores[['数学','英语','物理']].sum(axis=1)#①获取多列数据用两层中括号 ②计算总分并添加一列，sum默认统一列指标求和，axis=1是横向求和
scores['平均分']=(scores['总分']/3).round(1)

print('另一种方式')
scores['平均分']=(scores[['数学','英语','物理']].mean(axis=1)).round(3)
scores
```

```python
#按要求筛选
yx = (scores['数学']>90) & (scores['英语']>85)
result = scores[yx]
result
```

```python
scores.sort_values(['总分'],ascending=False)
```
